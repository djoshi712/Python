### Handling exception using try except

- put risky code inside try block
- write alternate solution in except block

```python
def m1():
    m2()
    print("hi")
    
def m2():   
    try:
        m3()
#          print Exception Information to the Console using as "e"
    except ZeroDivisionError as e:
        print("handling the code")
        print(type(e))
        
    print("bye")
    
def m3():
    print(10/0)
    
```

### finally block
excecuted always irrespective of whether exception is raised/handled or not

meant for cleanup code

finally won't excutes when we are using os._exit(0) >> PVM shutdown


```python
# if no exception
try:
    print("try")
except:
    print("except")
finally:
    print("finally")
```

```python
# if exception raised and handled
try:
    print("try")
    print(20/0)
except ZeroDivisionError:
    print("except")
finally:
    print("finally")
```

```python
# if exception raised and but handled
try:
    print("try")
    print(20/0)
except TypeErrot:
    print("except")
finally:
    print("finally")
```

```python
# finally always exceutes??
import os
try:
    print("try")
    os._exit(0)
except TypeError:
    print("except")
finally:
    print("finally")
```

```python
try:
    print("a")
    print("b")
    try:
        print("c")
        print(10/"10")
        print("e")
    except TypeError:
        print("handled")
    finally:
        print("f")
        
except ZeroDivisionError as msg:
    print(msg)
finally:
    print("finally block outside")
```
